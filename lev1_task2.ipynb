```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv("/content/Dataset .csv")
df=pd.DataFrame(data)
df
```

```python
most_restaurants = df['City'].value_counts().idxmax()
most_restaurants
```

```python
top_ten_cities = df['City'].value_counts().nlargest(10).index
df_top_ten_cities = df[df['City'].isin(top_ten_cities)]
```

```python
# Define the figure size
plt.figure(figsize=(12, 8))

# Create the count plot with the y variable assigned to hue and disable legend
sns.countplot(y='City', data=df_top_ten_cities, order=top_ten_cities, hue='City', palette='viridis', legend=False)

# Set the plot title and labels
plt.title('Top Ten Cities with the Most Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')

# Show the plot
plt.show()
```

```python
city = df.groupby('City')['Aggregate rating'].mean()
highest_avg_rating = city.idxmax()
highest_avg_rating
```

```python
average_ratings = df.groupby('City')['Aggregate rating'].mean()
top_ten_cities = average_ratings.sort_values(ascending=False)
top_ten_cities
```

```python
top_ten_cities = average_ratings.sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_ten_cities.sort_values().plot(kind='barh', color='blue')
plt.title('Average Rating of Top 10 Restaurants by City')
plt.xlabel('Average Rating')
plt.ylabel('City')
```
