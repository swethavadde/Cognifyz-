```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv("/content/Dataset .csv")
df=pd.DataFrame(data)
df
```

```python
# Convert 'Aggregate rating' column to numeric, with invalid values set to NaN
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Display the DataFrame to confirm changes
df.head()
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Define the figure size
plt.figure(figsize=(10, 6))

# Create the histogram
sns.histplot(df['Aggregate rating'], bins=range(1, 6), kde=False, color='pink', edgecolor='black')

# Set the plot title and axis labels
plt.title('Distribution of Aggregate Ratings Among Restaurants')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Count the occurrences of each aggregate rating
rating_counts = df['Aggregate rating'].value_counts()

# Define the figure size
plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))

# Set equal aspect ratio for a circular pie chart
plt.axis('equal')

# Set the title
plt.title('Distribution of Aggregate Ratings Among Restaurants')

# Show the plot
plt.show()

```
