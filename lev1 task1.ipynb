```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv("/content/Dataset .csv")
df=pd.DataFrame(data)
df
```

```python
Cuisines=df['Cuisines'].value_counts()
Cuisines.head(3)
```

```python
plt.figure(figsize=(10, 6))

# Plot the top 3 cuisines without specifying palette since hue is not assigned
sns.barplot(x=Cuisines.head(3).index, y=Cuisines.head(3).values)

# Add labels and title
plt.title('Top 3 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')

plt.show()

```

```python
split = df['Cuisines'].str.split(',').explode()
counts =split.value_counts()
percentages = (counts / len(df)) * 100
print("Percentage of Restaurants Serving Top Cuisine:")
print(percentages)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Cuisines.head(10).index, y=Cuisines.head(10).values, palette='viridis')
plt.title('Top Cuisines 10 and Their Percentages')
plt.show()
```
