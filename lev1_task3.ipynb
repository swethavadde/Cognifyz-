```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv("/content/Dataset .csv")
df=pd.DataFrame(data)
df
```

```python


# Define the figure size
plt.figure(figsize=(24, 12))

# Create the count plot and assign the 'x' variable to 'hue', and disable the legend
sns.countplot(x='Price range', data=df, hue='Price range', palette='viridis', legend=False)

# Set the plot title and labels
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python

price_range_counts = df['Price range'].value_counts()
percentage_per_price_range = (price_range_counts / len(df)) * 100
print("Percentage of Restaurants in Each Price Range:")
percentage_per_price_range
```
