```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
data=pd.read_csv("/content/Dataset .csv")
df=pd.DataFrame(data)
df
```

```python
data.info()
```

```python
data.describe()
```

```python
# Drop rows with missing values in specified columns
data = data.dropna(subset=['Price range', 'Has Table booking', 'Has Online delivery'])

# Display the updated DataFrame
data
```

```python
# Convert 'Yes'/'No' in 'Has Table booking' to True/False
data['Has Table booking'] = data['Has Table booking'].map({'Yes': True, 'No': False})

# Convert 'Yes'/'No' in 'Has Online delivery' to True/False
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': True, 'No': False})
```

```python
plt.figure(figsize=(10, 6))
heatmap = pd.crosstab(data['Price range'], [data['Has Table booking'], data['Has Online delivery']])
sns.heatmap(heatmap, annot=True, fmt='d', cmap='pink', cbar=True)
plt.title('Relationship between Price Range and Availability of Table Booking & Online Delivery')
```

```python
# Define the figure size
plt.figure(figsize=(12, 6))

# Create the count plot with custom colors
sns.countplot(x='Price range', hue='Has Online delivery', data=df, palette=['#3498db', '#e74c3c'])  # Blue for False, Red for True

# Set the plot title and labels
plt.title('Online Delivery Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')

# Show the plot
plt.show()
```
